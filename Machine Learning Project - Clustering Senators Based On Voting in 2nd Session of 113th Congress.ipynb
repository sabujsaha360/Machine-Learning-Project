# Clustering Senators Based  On Voting in 2nd Session of 113th Congress

```python
# Importing Data and Reading from data file
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
congress = pd.read_csv("C:/Users/sasaha/Documents/senate113_2nd session.csv")
```

```python
# Checking the first 5 rows of data set
congress.head()
```

```python
#Reading the bills columns
congress.columns[:-2]
```

```python
#Finding how many senators in each party
#Finding the average voting of each bill
print(congress["party"].value_counts())
print(congress.mean())
```

```python
#Finding the average voting of each bill based on part affliation
congress.groupby(['party'])[congress.columns[:-2]].mean()
```

```python
# To check with columns to select for K- Means clustering
congress.iloc[:, :-2].head()
```

```python
#Implementing K-Means clustering Algorithim by initiliazing two clusters since we have two main parties
#Estimate the difference of each senator from each clustering
import pandas as pd
from sklearn.cluster import KMeans
senator_cluster= KMeans(n_clusters=2, random_state=1)
senator_distances = senator_cluster.fit_transform(congress.iloc[:, :-2])
senator_distances[:,:2]
```

```python
# finding which cluster that  each senator falls 1st (0) or 2nd (1)
labels = senator_cluster.labels_
```

```python
# Taking seperate column to include labels in main dataset
congress['clustervote']= labels
congress.head()
```

```python
#How many Senators from each party ended up in each cluster
print(pd.crosstab(labels, congress["party"]))
```

Unexpected !!! Senators typically vote in accordance with how their political party votes, known as voting along party lines.However, both Democrat as well as Republican had deviated from their party decisions which are very surprising.Crosstable shows that democrat senators are far way more divided than Republican.

```python
# Lets find out who are outliers in democratic party
demo_exceptions = congress[(congress["clustervote"] == 1)& (congress["party"] == "D")]['name']
print(demo_exceptions)
```

```python
# Finding the name of top ten extreme senators along with the parties who are more aligned with their clusters than the other.
extremism = (senator_distances **3).sum(axis=1)
congress["extremism"] = extremism
congress.sort_values("extremism", inplace=True, ascending=False)
print(congress[['name','party']].head(10))
```

```python
# Sorting dataframe based on party affiliation to find the top 10 extreme senators of each party
top_extreme= congress.groupby(['party']).apply(lambda x: x.sort_values(["extremism"], ascending = False)).reset_index(drop=True)
# select top N rows within each continent
top_extreme.groupby('party')[["name","party",'extremism']].head(10)
```
